14. Read the image ('boy.jpg').

```python
img = cv2.imread('boy.jpg')
```

15. Adjust the brightness of the image.

```python
# Create a matrix of ones (with data type float64)
# matrix_ones = 
matrix_ones = np.ones(img.shape, dtype='float64') * 50
img_brighter = cv2.add(img, matrix_ones.astype('uint8'))
plt.imshow(cv2.cvtColor(img_brighter, cv2.COLOR_BGR2RGB))
plt.title("Brighter Image")
plt.axis('off')
plt.show()
```

16. Create brighter and darker images.

```python
#img_brighter = cv2.add(img, matrix)
#img_darker = cv2.subtract(img, matrix)
# YOUR CODE HERE
matrix = np.ones(img.shape, dtype='uint8') * 50
img_brighter = cv2.add(img, matrix)
img_darker = cv2.subtract(img, matrix)
```

17. Display the images (Original Image, Darker Image, Brighter Image).

```python
orig = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
bright = cv2.cvtColor(img_brighter, cv2.COLOR_BGR2RGB)
dark = cv2.cvtColor(img_darker, cv2.COLOR_BGR2RGB)

# Show all images
plt.figure(figsize=(10,4))

plt.subplot(1,3,1)
plt.imshow(orig)
plt.title("Original")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(bright)
plt.title("Brighter")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(dark)
plt.title("Darker")
plt.axis('off')

plt.show()
```

18. Modify the image contrast.

```python
# Create two higher contrast images using the 'scale' option with factors of 1.1 and 1.2 (without overflow fix)
# matrix1 = 
# matrix2 = 
# img_higher1 = 
# img_higher2 = 
matrix1 = np.ones(img.shape, dtype='float32') * 1.1
matrix2 = np.ones(img.shape, dtype='float32') * 1.2
img_higher1 = img * matrix1
img_higher2 = img * matrix2
img_higher1 = np.clip(img_higher1, 0, 255).astype('uint8')
img_higher2 = np.clip(img_higher2, 0, 255).astype('uint8')
```

19. Display the images (Original, Lower Contrast, Higher Contrast).

```python
orig = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
h1 = cv2.cvtColor(img_higher1, cv2.COLOR_BGR2RGB)
h2 = cv2.cvtColor(img_higher2, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,4))
plt.subplot(1,3,1)
plt.imshow(orig)
plt.title("Original")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(h1)
plt.title("Contrast 1.1")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(h2) 
plt.title("Contrast 1.2")
plt.axis('off')
plt.show()
```

20. Split the image (boy.jpg) into the B,G,R components & Display the channels.

```python
b, g, r = cv2.split(img)


plt.figure(figsize=(10,4))

plt.subplot(1,3,1)
plt.imshow(b, cmap='gray')
plt.title("Blue Channel")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(g, cmap='gray')
plt.title("Green Channel")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(r, cmap='gray')
plt.title("Red Channel")
plt.axis('off')

plt.show()
```

21. _merged the R, G, B , display along with orginal image 

```python
merged = cv2.merge([b, g, r])
orig_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
merged_rgb = cv2.cvtColor(merged, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(orig_rgb)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(merged_rgb)
plt.title("Merged Image (BGR)")
plt.axis('off')

plt.show()
```

22. Split the image into the H, S, V components &  Display the channels.

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)

plt.figure(figsize=(10,4))

plt.subplot(1,3,1)
plt.imshow(h, cmap='gray')
plt.title("Hue Channel")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(s, cmap='gray')
plt.title("Saturation Channel")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(v, cmap='gray')
plt.title("Value Channel")
plt.axis('off')

plt.show()
```

23. _merged the H, S, V, display along with orginal image 

```python
merged_hsv = cv2.merge([h, s, v])
merged_rgb = cv2.cvtColor(merged_hsv, cv2.COLOR_HSV2RGB)
orig_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(orig_rgb)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(merged_rgb)
plt.title("Merged HSV Image")
plt.axis('off')

plt.show()
```
